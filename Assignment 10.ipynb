```python
class NQueens:
    def __init__(self, n):
        self.n = n
        self.board = [[0] * n for _ in range(n)]
        self.solution = []

    def solve(self):
        if self.backtrack(0):
            print("Solution found:")
            self.print_board()
        else:
            print("No solution found.")

    def backtrack(self, col):
        if col == self.n:
            # Found a valid solution
            self.solution = [row[:] for row in self.board]
            return True

        for row in range(self.n):
            if self.is_safe(row, col):
                self.board[row][col] = 1  # Place a queen in the cell

                if self.backtrack(col + 1):
                    return True

                self.board[row][col] = 0  # Backtrack and remove the queen

        return False

    def is_safe(self, row, col):
        # Check if a queen can be placed at the given position without attacking any other queens

        # Check the row
        for i in range(col):
            if self.board[row][i] == 1:
                return False

        # Check the upper diagonal
        i = row
        j = col
        while i >= 0 and j >= 0:
            if self.board[i][j] == 1:
                return False
            i -= 1
            j -= 1

        # Check the lower diagonal
        i = row
        j = col
        while i < self.n and j >= 0:
            if self.board[i][j] == 1:
                return False
            i += 1
            j -= 1

        return True

    def print_board(self):
        for row in self.solution:
            print(" ".join(str(cell) for cell in row))


# User input to get the number of queens
print("Enter the number of queens:")
n = int(input())

# Create an instance of the NQueens class and solve the problem
queens = NQueens(n)
queens.solve()
```
